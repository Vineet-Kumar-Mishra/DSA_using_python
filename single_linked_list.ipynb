```python
class Node():
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next
```

```python
class Linked_list():
    def __init__(self):
        
        # take an empty linkedlist and make head as None
        self.head = None
        
    # 1st function is to add element at beginning
    def insert_at_beginning(self, data):
        
        # if it's an empty linkedlist, simply make the head as node with the data
        if self.head is None:
            self.head = Node(data)
            return
        
        # if head is present then make node as new head with pointing to the previous head
        self.head = Node(data, self.head)
    
    # 2nd function is to print the list
    def print(self):
        
        # if has no head then print empty
        if self.head is None:
            print("Empty linked list")
            return
        
        # if head present then iterate over all values starting from head 
        # update iterator to the next element using next
        itr = self.head
        string = ""
        while itr:
            string += str(itr.data)+"-->" if itr.next else str(itr.data)
            itr = itr.next
        print(string)
        
    # 3rd funtion is to append at the end
    def insert_at_end(self, data):
        
        # if head empty then make new node with pointing towards None
        if self.head is None:
            self.head = Node(data, None)
            return
        itr = self.head
        # check if next is present so we stop at last element (i-1)
        while itr.next:
            itr = itr.next
        itr.next = Node(data,None)
        
    # 4th function to get length of list
    def len(self):
        count = 0
        itr = self.head
        while itr:
            count+=1
            itr = itr.next
        return count
    
    # 5th function to append at index
    def insert_at_index(self, index, data):
        if index<0 or index>self.len():
            raise Exception("Invalid index")
        
        # if 0 index then use insert at beginning function
        if index==0:
            self.insert_at_beginning(data)
        
        # we will count the elements traversed and match with index
        count = 0
        itr = self.head
        while itr:
            if count==index-1:
                node = Node(data,itr.next)
                itr.next = node
                break
            itr = itr.next
            count +=1
            
    # 6th function to remove at index
    def remove_At_index(self,index):
        if index<0 or index>self.len():
            raise Exception("Invalid index")
        
        count = 0
        itr = self.head
        while itr:
            if count==index-1:
                itr.next = itr.next.next
                break
            itr = itr.next
            count+=1
            
    # 7th function to insert a number after a number
    def insert_after_data(self, data_after, data):
        itr = self.head
        while itr:
            if itr.data==data_after:
                itr.next = Node(data, itr.next)
                break
            itr = itr.next
    # 8th functio to remove after a number
    def remove_after_data(self, data_after):
        itr = self.head
        while itr.next:
            if itr.data==data_after:
                itr.next = itr.next.next
                break
            itr = itr.next
    # 9th function to insert python list into linkedlist
    def list_to_linkedlist(self, list_python):
        self.head = None
        for data in list_python:
            self.insert_at_end(data)
```

```python
l1 = Linked_list()
l1.insert_at_beginning(25)
l1.insert_at_beginning(30)
l1.insert_at_end(11)
l1.insert_at_end(21)
l1.insert_at_end(40)
l1.insert_at_end(45)
l1.print()
print(l1.len())
l1.insert_at_index(2,80)
l1.print()
l1.insert_at_index(0,90)
l1.print()
l1.remove_At_index(3)
l1.print()
l1.insert_after_data(21,89)
l1.print()
l1.remove_after_data(11)
l1.print()
```

```python
l2 = Linked_list()
l2.list_to_linkedlist([25,80,90,60])
l2.print()
```
